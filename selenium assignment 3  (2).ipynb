```python
url = []
for i in driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']"):
    url.append(i.get_attribute("href"))
```

```python
len(url)
```

```python
model_ = []
```

```python
for i in url:
    driver.get(i)
    time.sleep(3)
    try:
        model = driver.find_element_by_xpath("//span[@class='a-size-large product-title-word-break']")
        model_.append(model.text.replace("/n","new line"))
    except:
        model_.append('-')
```

```python
model_
```

```python
tim_ = []
```

```python
for i in url:
    driver.get(i)
    time.sleep(3)
    try:
        tim = driver.find_element_by_xpath("//div[@class='a-section a-spacing-top-mini']")
        tim_.append(tim.text.replace("/n","new line"))
    except:
        tim_.append('-')
```

```python
tim_
```

```python
price_ = []
```

```python
for i in url:
    driver.get(i)
    time.sleep(1)
    try:
        price = driver.find_element_by_xpath("//span[@class='a-size-medium a-color-price priceBlockBuyingPriceString']")
        price_.append(price.text.replace("/n","new line"))
    except:
        price_.append('-')
```

```python
price_
```

```python
import pandas as pd
```

```python
gui = pd.DataFrame({})
```

```python
gui["model"]=model_[0:65]
gui["delivery"]=tim_[0:65]
gui["price"]=price_[0:65]
```
